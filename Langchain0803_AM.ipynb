## Problem Statement
In training sessions, different learners want explanations based on the subject and topic they are learning. For example, one learner may want to understand Large Language Models, while another may ask about Machine Learning or Blockchain.

A trainer cannot manually rewrite explanations for every topic and every domain again and again.

So, we need a simple AI-based solution that can:

take the domain as input,

take the topic as input,

and generate a beginner-friendly explanation automatically.

This code solves that problem by using LangChain, OpenAI, and a prompt template to create a smart training assistant that gives simple explanations based on the user’s input.

```python
!pip install -q langchain langchain-openai

import os
from google.colab import userdata
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are an expert corporate trainer specializing in {domain}."),
    ("human", "Explain {topic} in simple terms for a beginner.")
])

chain = prompt | llm | StrOutputParser()

response = chain.invoke({
    "domain": "AI & Technology",
    "topic": "Large Language Models"
})

print(response)
```

## 5. Problem Statement
In training sessions or AI assistants, users often ask multiple questions one after another. A normal chatbot may answer each question separately and forget previous conversation details, such as the user’s name or earlier questions.

This creates a problem because the assistant cannot maintain context or memory during a conversation.

So, we need a smart AI assistant that can:

remember previous messages,

understand the conversation context,

and respond intelligently based on earlier interactions.

This code solves that problem by using LangChain with chat history, allowing the AI assistant to remember previous user inputs and continue the conversation naturally.


```python
!pip install -q langchain langchain-openai

import os
from google.colab import userdata
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.7)

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful corporate training assistant."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

chain = prompt | llm

# Store chat history
chat_history = []

def chat(user_input):
    response = chain.invoke({
        "input": user_input,
        "chat_history": chat_history
    })
    chat_history.append(HumanMessage(content=user_input))
    chat_history.append(AIMessage(content=response.content))
    print(f"You: {user_input}")
    print(f"Bot: {response.content}\n")

# Simulate a multi-turn conversation
chat("Hi, my name is Arun.")
chat("What topics are covered in an Agentic AI workshop?")
chat("Can you remind me what my name is?")
#print (chat_history)
```

## 6. Problem Statement
AI models have a lot of knowledge, but sometimes they do not have the latest or real-time information, especially about new trends or recent developments.

For example, if a user asks:

“What are the latest trends in Agentic AI in 2025?”

The AI model alone may not always give the most updated answer because it may not have access to live internet data.

So, we need a solution where the AI can:

search the internet for latest information,

retrieve the relevant results,

and use that information to generate a more accurate and updated answer.

This code solves that problem by connecting the AI model with a web search tool (DuckDuckGo) using LangChain. The AI first performs a search, gets the latest results, and then generates a final answer based on that information.

```python
!pip install -q langchain langchain-openai langchain-community ddgs

import os
from google.colab import userdata
from langchain_openai import ChatOpenAI
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.messages import HumanMessage, ToolMessage

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
search = DuckDuckGoSearchRun()
llm_with_tools = llm.bind_tools([search])

user_input = "What are the latest trends in Agentic AI in 2025?"
response = llm_with_tools.invoke(user_input)

search_result = search.run(response.tool_calls[0]["args"]["query"])

messages = [
    HumanMessage(content=user_input),
    response,
    ToolMessage(content=search_result, tool_call_id=response.tool_calls[0]["id"])
]

final = llm.invoke(messages)
print(final.content)
```

## 7. Problem Statement
AI models can answer many questions, but sometimes they may not provide detailed factual information from reliable sources such as encyclopedias.

For example, if a user asks:

“What is the history of Artificial Intelligence?”

The AI may generate a general answer, but it may not always include accurate historical details from trusted knowledge sources.

So, we need a system where the AI can:

search reliable knowledge sources like Wikipedia,

retrieve relevant information,

and use that information to generate a more accurate and informative answer.

This code solves that problem by connecting the AI model with a Wikipedia search tool using LangChain. The AI first queries Wikipedia, retrieves the information, and then generates a final answer based on that data.

```python
!pip install -q langchain langchain-openai langchain-community wikipedia

import os
from google.colab import userdata
from langchain_openai import ChatOpenAI
from langchain_community.tools import WikipediaQueryRun
from langchain_community.utilities import WikipediaAPIWrapper
from langchain_core.messages import HumanMessage, ToolMessage

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
wiki = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())
llm_with_tools = llm.bind_tools([wiki])

user_input = "What is the history of Artificial Intelligence?"
response = llm_with_tools.invoke(user_input)

wiki_result = wiki.run(response.tool_calls[0]["args"]["query"])

messages = [
    HumanMessage(content=user_input),
    response,
    ToolMessage(content=wiki_result, tool_call_id=response.tool_calls[0]["id"])
]

final = llm.invoke(messages)
print(final.content)
```

## 8. Problem Statement
In many organizations, important information is stored in PDF documents such as HR policies, manuals, reports, or guidelines. When employees want to find specific information, they usually have to read the entire document manually, which can take a lot of time.

For example, if an employee asks:

“Summarize the HR policy document”

They may need to go through many pages to find the answer.

So, we need a smart AI system that can:

read a PDF document,

understand the content,

store the information in a searchable format,

and quickly answer questions based on that document.

This code solves that problem by using LangChain, embeddings, and a FAISS vector database to create a RAG (Retrieval-Augmented Generation) system. The system reads the uploaded PDF, stores its content as embeddings, retrieves relevant information, and generates an answer based only on the document content.

```python
!pip install -q langchain langchain-openai langchain-community pypdf langchain-text-splitters faiss-cpu

import os
from google.colab import userdata, files
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

# Step 1: Upload PDF
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Step 2: Load & Split
chunks = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=50).split_documents(PyPDFLoader(file_name).load())

# Step 3: Store in FAISS & Save to disk
vectorstore = FAISS.from_documents(chunks, OpenAIEmbeddings())
vectorstore.save_local("./faiss_db")
print(f"✅ Saved to ./faiss_db")
print(f"Total chunks: {vectorstore.index.ntotal}")

retriever = vectorstore.as_retriever()

# Step 4: RAG Chain
prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer using only the context below.\n\n{context}"),
    ("human", "{question}")
])

chain = (
    {"context": retriever, "question": lambda x: x}
    | prompt
    | ChatOpenAI(model="gpt-4o-mini", temperature=0)
    | StrOutputParser()
)

# Step 5: Ask a question
print(chain.invoke("Summarize the HR policy document"))
```
